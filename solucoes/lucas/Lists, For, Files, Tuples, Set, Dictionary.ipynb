## Problem 25: 
Provide an implementation for `zip` function using `list comprehensions`.
```
>>> zip([1, 2, 3], ["a", "b", "c"])
    [(1, "a"), (2, "b"), (3, "c")]
```

````python
a=[1, 2, 3]
b=["a", "b", "c", "d", "e"]

def zip(a,b):
    
    zip_list=[]
    
    for i in range(min(len(a),len(b))):
        zip_list.append((a[i],b[i]))
        
    return zip_list

zip(a,b)
````

````python
#list comprehension
a=[1, 2, 3]
b=["a", "b", "c"]

def zip(a,b):
    return [(a[i], b[i]) for i in range(min(len(a),len(b)))]

zip(a,b)
````
